```python
import pickle
with open("models/vectorizer.pkl","rb") as s:
    vectorizer = pickle.load(s)
with open("models/model.pkl","rb") as f:
    model = pickle.load(f)
```

```python
def predict(text):
    transform = vectorizer.transform([text]).toarray()
    return "spam" if model.predict(transform) == 0 else "not spam"
```

```python
predict("Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat...'")
```
